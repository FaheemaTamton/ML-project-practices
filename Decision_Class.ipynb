***About Dataset:**
The "Baseball Player Metrics" dataset comprises comprehensive statistical records for a diverse range of baseball players. It encompasses essential performance metrics such as games played, at-bats, runs scored, hits, doubles, triples, home runs, and runs batted in (RBI) for each player. With data spanning multiple years and including unique players, this dataset offers insights into individual player performance and allows for comparative analyses across different eras and player styles. Researchers, analysts, and enthusiasts can utilize this dataset to explore trends, assess player contributions, and gain a deeper understanding of the dynamics of baseball over time. Whether studying the evolution of the game or evaluating the impact of specific players, this dataset serves as a valuable resource for baseball-related analysis and research endeavors.*

*Player: The name of the baseball player.
YRS: The number of years the player has been active in professional baseball.
G: The total number of games played by the player.
AB: The total number of at-bats (plate appearances) by the player.
R: The total number of runs scored by the player.
H: The total number of hits recorded by the player.
2B: The total number of doubles hit by the player.
3B: The total number of triples hit by the player.
HR: The total number of home runs hit by the player.
RBI: The total number of runs batted in (RBIs) by the player.*

```python
df = df.drop(columns = ['PLAYER', 'CS'])
```

```python
x =df.iloc[:, 0:13]
```

```python
y = df.iloc[:, 13]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=17, test_size = 0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()#object creation 
```

```python
dtc.get_params()
```

```python
#Traning
dtc.fit(x_train, y_train)
```

```python
#prediction
y_pred = dtc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred)) #prints True-positive, T-N, F-P, F-N
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
