# RANDOM FOREST CLASSIFIER 

### **Digits dataset from sklearn**

```python
#Data Manipulation
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
#visualization
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
plt.show()
```

```python
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
digits.target
```

```python
#creating a new col 'target'
df['target']=digits.target
df.head()
```

```python
df[0:12]
```

### Train and the model prediction

```python
x = df.drop('target', axis = 'columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
len(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier#(use ensemble because : using multiple algo to predict the outcome)
model = RandomForestClassifier(n_estimators=40)#number of trees
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
#confusion metrix to see the ditribution of errors and where my model perform well vs poor, 
y_predicted = model.predict(x_test)
```

### **Confusion matrix**

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('Truth')
plt.show() #visualising confusion matrix
```
